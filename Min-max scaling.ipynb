```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# read csv as df
df = pd.read_csv("data_modelling.csv")
```

```python
df.head()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']) # Extract numeric columns
numeric_cols = numeric_cols.drop(['Record ID'], axis=1)       # drop record id
scaler = MinMaxScaler()                                       # instantiate MinMaxScaler
df[numeric_cols.columns] = scaler.fit_transform(numeric_cols) # apply MinMaxScaler
df.describe()
```

```python
df.describe()
```

```python
# save as csv
df.to_csv("data_normalized2.csv")
```

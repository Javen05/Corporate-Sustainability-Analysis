```python
import pandas as pd
import numpy as np
```

```python
# read csv as df
df = pd.read_csv("data_modelling.csv")
```

```python
df.head()
```

```python
numeric_cols = df.select_dtypes(include=[np.number])                                                # Extract numeric columns
numeric_cols = numeric_cols.drop(['Record ID'], axis=1)
df[numeric_cols.columns] = numeric_cols.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x))) # min-max normalization
df.head()
```

```python
# save as csv
df.to_csv("data_normalized.csv")
```
